```python
path = "data/person"
face_encoder_map = {}
fn: str
for fn in os.listdir(path):
    name = fn.removesuffix(".png")
    name = fn.removesuffix(".jpg")
    print(fn)
    face_encoder_map[fn] = face_encodings(
        face_recognition.load_image_file(path + "/" + fn))[0]
    
    
```

```python
cap = cv2.VideoCapture(0)
```

```python
frames = []
```

```python
while (1):
    ret, frame = cap.read()
    # frames.append(frame)
    
    # 发现在视频帧所有的脸和face_enqcodings

        # 读取当前帧

        
    face_location = face_recognition.face_locations(frame)
    face_encoders = face_recognition.face_encodings(frame, face_location)[0]
    (top, right, bottom, left) = face_location[0]
    
    for i, v in enumerate(face_encoder_map):
            match = face_recognition.compare_faces(
                [face_encoder_map[v]], face_encoders, tolerance=0.5)
            name = "Unknown"
            if match[0]:
                name = v
                break
            else:
                print(f"not match {v}")
        #     cv2.rectangle(frame, (left, top), (right, bottom), (0, 0, 255), 2)
    #     # 画出一个带名字的标签，放在框下
    #     cv2.rectangle(frame, (left, bottom - 35), (right, bottom), (0, 0, 255),
    #                   cv2.FILLED)
    font = cv2.FONT_HERSHEY_DUPLEX
    #     cv2.putText(frame, name, (left + 6, bottom - 6), font, 1.0,
    #                 (255, 255, 255), 1)
    cv2.rectangle(frame, (left, top), (right, bottom), (0, 0, 255), 2)
# 画出一个带名字的标签，放在框下
    cv2.rectangle(frame, (left, bottom - 35), (right, bottom), (0, 0, 255),
              cv2.FILLED)
    cv2.putText(frame, name, (left + 6, bottom - 6), font, 1.0,
            (255, 255, 255), 1)
    cv2.imshow('frame', frame)
    if cv2.waitKey(10) == 27:                           #按键盘的ESC键可退出，同时停顿一微秒
        break
    if cv2.getWindowProperty('frame',cv2.WND_PROP_AUTOSIZE) < 1:        #用鼠标点击窗口退出键实现退出循环
        break
cap.release()
cv2.destroyAllWindows()
```
